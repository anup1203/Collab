Q1. Which state is having highest honey production through out all these  years.(represent it using a bar graph)

```python
higest_honey = df.groupby('state')['production'].sum().sort_values(ascending= False).reset_index()
```

```python
higest_honey


```

```python
total_production_by_state = df.groupby('state')['production'].sum().sort_values(ascending = False).reset_index()
```

```python
plt.figure(figsize=(10,10))
sns.barplot(data = total_production_by_state , x= 'production', y = 'state',palette='hsv')
plt.show()
```

# NorthDakota is the highest honey production state

Q2. Which sate is having the highest and lowest yield_per_colony

```python
yield_high_low = df.groupby('state')['yield_per_colony'].sum().sort_values(ascending= False).reset_index()
```
